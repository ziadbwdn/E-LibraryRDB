```python
# 1st step: import library
from faker import Faker
import pandas as pd
import random
from datetime import datetime, timedelta
```

Make Each Table Required

```python
# Define Libraries Table
def create_libraries(library_name, library_address):
    """
    create a DataFrame for libraries.

    Parameters:
    - library_names (list): list of library names, it has 8 names.
    - library_address (list): list of library address, it has 8 address.

    Returns:
    - pd.DataFrame: DataFrame containing library information such as library_id and library_name.
    """
    
    libraries = {
        "library_id": [i + 1 for i in range(len(library_name))],
        "library_name": library_name,
        "library_address": library_address
    }

    libraries_df = pd.DataFrame(libraries)
    return libraries_df

library_name = ["Wakanda Library", 
                "Pustaka Konoha", 
                "Pohon Pengetahuan Pulau Ohara",
                "Microlibrary Warak Kayu",
                "C2O Library",
                "Kineruku Library",
                "Baca Di Tebet",
                "Taman Baca Amin Perpustakaan Kontainer",
                ]
    
library_address = ["Jl Margomulyo Permai 6-8 Kav 2 D-G Surabaya Jawa Timur", 
                   "Jl Cimandiri 34-38 Bandung Jawa Barat",
                   "Jl Pulo Gebang Raya 25 Jakarta DKI Jakarta",
                   "Jl Cirebon 83 Medan Sumatera Utara",
                   "Jl Manyar Jaya 18-20 Surabaya Jawa Timur",
                   "Jl Batutulis Raya 26 Jakarta DKI Jakarta",
                   "Jl Terogong Raya 16 Jakarta DKI Jakarta",
                   "Jl Ir H Juanda 377 Bandung Jawa Barat"
                  ]
libraries_table = create_libraries(library_name, library_address)
libraries_table
```

```python
# Define Categories Table
def create_categories(category_names):
    """
    create a DataFrame for categories.

    Parameters:
    - category_names (list): list of each category. there are 12 of category names.

    Returns:
    - pd.DataFrame: DataFrame containing category information such as category_id and category_name.
    """
    categories = {
        "category_id": [i + 1 for i in range(len(category_names))],
        "category_name": category_names
    }

    categories_df = pd.DataFrame(categories)
    return categories_df

#making list of category names
category_names = ["Fiction - Action and Adventure", 
                  "Fiction - Historical Fiction", 
                  "Fiction - Horror", 
                  "Fiction - Young Adult",
                  "Fiction - Romance",
                  "Fiction - Contemporary",
                  "Nonfiction - History", 
                  "Nonfiction - Travel", 
                  "Nonfiction - Self-Help",
                  "Nonfiction - Science & Technology",
                  "Nonfiction - Biography", 
                  "Nonfiction - Social Sciences",
                 ]

categories_table = create_categories(category_names)
categories_table
```

```python
# Define Publishers Table
def create_publishers(publisher_name):
    """
    create a DataFrame for publishers and make a mapping between publisher_name and publisher_id.

    Parameters:
    - publisher_names (list): list of each publisher. there are 14 of category names.

    Returns:
    - pd.DataFrame: DataFrame containing publisher information.
    """
    publishers = {
        "publisher_id": [i + 1 for i in range(len(publisher_name))],
        "publisher_name": publisher_name
    }
    
    publishers_df = pd.DataFrame(publishers)
    return publishers_df

# making list of publisher names
publisher_name = ["Penerbit Kompas Pustaka Gramedia",
                   "Penerbit Gramedia Pustaka Utama",
                   "Elex Media Komputindo",
                   "Penerbit Mizan",
                   "Penguin Random House",
                   "Penerbit Gagas Media", 
                   "Penerbit Bentang Pustaka", 
                   "MacMillan Publishing Group",
                   "Penerbit Erlangga",
                   "Penerbit Tiga Serangkai", 
                   "Verso Books", 
                   "Penerbit Deepublish", 
                   "Pustaka Obor", 
                   "Simon & Schuster"
                  ]
#create publishers DataFrame
publishers_table =  create_publishers(publisher_name)
publishers_table
```

```python
# Define Authors Table
def create_authors(n_authors):
    """
    create a DataFrame for authors and make a mapping between author_name and author_id.

    Parameters:
    - n_authors (int): number of authors information to create.

    Returns:
    - pd.DataFrame: DataFrame containing author information such as author id and author name.
    """
    authors = []
    for a in range(n_authors):
        authors.append({
            'author_id': fake.unique.random_int(min=1, max=9999),
            'author_name': fake.name()
        })
    
    authors_df = pd.DataFrame(authors)
    return authors_df
    
#create authors DataFrame and Mapping
authors_table = create_authors(800)
authors_table
```

```python
# Define Members Table
def create_members(n_members):
    """
    create a DataFrame for customers using faker library.

    Parameters:
    - n_members (int): number of customers.

    Returns:
    - pd.DataFrame: DataFrame containing members information.
    """
    members = {
        "member_id": [i + 1 for i in range(n_members)],
        "member_name": [],
        "member_address": [],
        "email": []
    }
    for i in range(n_members):
        member_name = fake.name()
        member_address = fake.address()
        email = f"{member_name.lower().replace(" ", "")}.{random.randint(1,100)}@xmail.com"

        members["member_name"].append(member_name)
        members["member_address"].append(member_address)
        members["email"].append(email)

    members_df = pd.DataFrame(members)
    return members_df

n_members = 1000
members_table = create_members(n_members)
members_table
```

```python
# Define Users Table
def create_users(n_users, members_table):
    """
    create a DataFrame for customers using faker library.

    Parameters:
    - n_customers (int): number of customers.

    Returns:
    - pd.DataFrame: DataFrame containing customer information.
    """
    users = {
        "user_id": [fake.unique.random_int(min=1, max=2000) for _ in range(n_users)],
        "user_password": [fake.password() for _ in range(n_users)],
        "member_id": [random.choice(members_table['member_id']) for _ in range(n_users)],
        "phone_number": [fake.unique.phone_number().replace("", "") for _ in range(n_users)]
    }

    users_df = pd.DataFrame(users)
    return users_df

n_users = 1500
users_table = create_users(n_users,users_table)
users_table
```

```python
# Define Books Table
def create_books(n_books, categories_table, authors_table, publishers_table):
    """
    create a DataFrame for books.

    Parameters:
    - n_books (int): number of book.
    - categories_table: DataFrame containing category information.
    - authors_table: DataFrame containing author information.
    - publishers_table (dict): DataFrame containing publisher information.

    Returns:
    - pd.DataFrame: DataFrame containing book information.
    """
    books = {
        "book_id": [i + 1 for i in range(n_books)],
        "category_id": [],
        "book_title": [],
        "author_id": [],
        "publisher_id": [],
        "book_language": [],
        "book_type": []
    }

    for i in range(n_books):
        first_article = random.choice(['The', 'A', 'An'])
        subject = fake.word().title()
        additional_word = fake.word().title()
        book_series = f" Book {random.randint(2, 5)}" if random.choice([True, False]) else ""
        book_title = f"{first_article} {subject} {additional_word} {book_series}".strip()
        
        books["category_id"].append(random.choice(categories_table['category_id']))
        books["book_title"].append(book_title)
        books["author_id"].append(random.choice(authors_table['author_id']))
        books["publisher_id"].append(random.choice(publishers_table['publisher_id']))
        books["book_language"].append(random.choices(['Bahasa', 'English'], weights=[0.1, 0.9])[0])
        books["book_type"].append(random.choice(['physical book','e-book','audiobook']))

    books_df = pd.DataFrame(books)
    return books_df

n_books = 10000
books_table = create_books(n_books, categories_table, authors_table, publishers_table)
books_table
```

```python
# Define Library Collections Table
def create_library_collections(n_library_collections, libraries_table, books_table):
    """
    create a DataFrame for library books.

    Parameters:
    - n_collections (int): number of library_collections.
    - libraries_df (pd.DataFrame): DataFrame containing library information.
    - books_df (pd.DataFrame): DataFrame containing book information.

    Returns:
    - pd.DataFrame: DataFrame containing library book information.
    """
    library_collections = {
        "library_collection_id": [i + 1 for i in range(n_library_collections)],
        "library_id": [random.choice(libraries_table['library_id']) for i in range(n_library_collections)],
        "book_id": [random.choice(books_table['book_id']) for i in range(n_library_collections)],
        "quantity_available": [random.randint(1, 4) for i in range(n_library_collections)]
    }

    df = pd.DataFrame(library_collections)
    return df

n_library_collections = n_books
library_collections_table = create_library_collections(n_library_collections, libraries_table, books_table)
library_collections_table
```

```python
# Define Library Collections Table
def create_borrows(n_borrows, users_table, library_collections_table):
    """
    create a DataFrame for borrows.

    Parameters:
    - n_borrows (int): number of borrows.
    - users_table (pd.DataFrame): DataFrame containing user information.
    - library_collections_table (pd.DataFrame): DataFrame containing library collection information.

    Returns:
    - pd.DataFrame: DataFrame containing borrowing information.
    """
    borrows = {
        "borrow_id": [i + 1 for i in range(n_borrows)],
        "user_id": [random.choice(users_table["user_id"]) for i in range(n_borrows)],
        "library_collection_id": [random.choice(library_collections_table["library_collection_id"]) for i in range(n_borrows)],
        "borrow_date": [],
        "due_date": [],
        "return_date": [],
        "return_status": []
    }

    # create empty list for start date, due date, and return date
    borrow_date_data = []
    due_date_data = []
    return_date_data = []

    # count of loans book per customer
    user_borrows_count = {}

    for i in range(n_borrows):
        user_id = borrows["user_id"][i]

        # count number of books by customer loans that hasn't been returning
        borrows_count = user_borrows_count.get(user_id, 0)

        # if the customer loans less than 2 books, allow to get another
        if borrows_count < 2:
            library_collection_id = random.choice(library_collections_table["library_collection_id"])
            borrows["library_collection_id"][i] = library_collection_id
            user_borrows_count[user_id] = borrows_count + 1

        # generate start date
        borrow_date = fake.date_time_between(start_date=datetime(2023, 1, 1), end_date=datetime.now())
        borrow_date_data.append(borrow_date)

        # generate return date
        return_date = (
            None if borrow_date is None
            else (borrow_date + timedelta(days=random.randint(1, 30))) if (datetime.now() - borrow_date).days < 14
            else borrow_date + timedelta(days=14)
        )
        return_date_data.append(return_date)

    # return_status are updated based on return_time
    borrows["return_status"] = [return_date is not None and return_date <= datetime.now() for return_date in return_date_data]

    # append to start_date and return_date columns
    borrows["borrow_date"] = borrow_date_data
    borrows["due_date"] = borrow_date + pd.Timedelta(days=14)
    borrows["return_date"] = return_date_data

    # create DataFrame
    df = pd.DataFrame(borrows)
    return df

n_borrows = 1200
borrows_table = create_borrows(n_borrows, users_table, library_collections_table)
borrows_table
```

```python
# Define holds table
def create_holds(n_holds, users_table, library_collections_table):
    """
    create a DataFrame for holds.

    Parameters:
    - n_holds (int): number of holds.
    - users_table (pd.DataFrame): DataFrame containing users information.
    - library_collections_table (pd.DataFrame): DataFrame containing library collections information.

    Returns:
    - pd.DataFrame: DataFrame containing hold information.
    """
    holds_table = {
        "hold_id": [i + 1 for i in range(n_holds)],
        "user_id": [random.choice(users_table["user_id"]) for i in range(n_holds)],
        "library_collection_id": [random.choice(library_collections_table["library_collection_id"]) for i in range(n_holds)],
        "hold_start": [],
        "hold_end": []
    }

    # create empty lists for hold time and end time
    hold_start_data = []
    hold_end_data = []

    # count of holds book per customer
    user_holds_count = {}

    for i in range(n_holds):
        user_id = holds_table["user_id"][i]

        # count number of books by customer holds that hasn't been returning
        holds_count = user_holds_count.get(user_id, 0)

        # if the customer holds less than 2 books, allow to get another
        if holds_count < 2:
            library_collection_id = random.choice(library_collections_table["library_collection_id"])
            holds_table["library_collection_id"][i] = library_collection_id
            user_holds_count[user_id] = holds_count + 1

        # Generate hold time
        hold_start = fake.date_time_between(start_date=datetime(2023, 1, 1), end_date=datetime.now())
        hold_start_data.append(hold_start)

        # Generate holds end (50% have return time 1 week after hold time)
        if random.random() < 0.5:
            hold_end = hold_start + timedelta(days=7)
        else:
            hold_end = None
        hold_end_data.append(hold_end)

    # Update holds_time and holds_end columns
    holds_table["hold_start"] = hold_start_data
    holds_table["hold_end"] = hold_end_data

    # Create DataFrame
    df = pd.DataFrame(holds_table)
    return df

n_holds = 800
holds_table = create_holds(n_holds, users_table, library_collections_table)
holds_table
```
